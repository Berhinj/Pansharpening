```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dft(x):
    x = np.asarray(x, dtype=float)
    N = x.shape[0]
    n = np.arange(N)
    k = n.reshape((N, 1))
    M = np.exp(-2j * np.pi * k * n / N)
    return np.dot(M, x)
```

```python
bit = 64
cycle = 64 * bit

x = np.exp(2j * np.pi * np.arange(cycle) / bit) #np.fft.fft()
plt.plot(x)
plt.show()
```
